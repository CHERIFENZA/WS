```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
code_source = requests.get ('https://fr.wikipedia.org/wiki/Python_(langage)'). content
code_source 
```

```python
soup = BeautifulSoup(code_source,'html')
```

```python
table = soup.select_one('table.infobox_v2')
```

```python
trs = table.select('tr')
```

```python
D = {} 
for tr in trs : 
    th = tr.select_one ('th')
    td = tr.select_one ('td')
    if th is None :
        continue

    D[td.text.strip ()] = td.text.strip() 
```

```python
pd.DataFrame ([D])
```
